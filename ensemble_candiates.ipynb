# Candidates for stacking ensemble

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.metrics import f1_score
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB, GaussianNB
from lightgbm import LGBMClassifier

from utilities import task1_load_cases, task2_load_cases, task3_load_cases
```

```python
def evaluate_algorithms(clfs, x_train, y_train, x_val, y_val, title):

    names = []
    scores = []

    for i, clf in enumerate(clfs):
        names.append(str(type(clf)).split(".")[-1][:-2])
        print(f'Fitting {i+1}/{len(clfs)}: {names[i]}')
        clf.fit(x_train, y_train)
        preds = clf.predict(x_val)
        scores.append(f1_score(y_val, preds, average='macro'))

    df = pd.DataFrame({"Score": scores, "Algorithm": names})
    g = sns.barplot(x="Score", y="Algorithm", data=df, orient="h")
    g.set_xlabel("Macro F1 score")
    g.set_title(title)
    plt.savefig('./plots/' + title +'.png', dpi=400)
    plt.show()
```

# Task 1

```python
classifiers = [
    SVC(),
    DecisionTreeClassifier(),
    AdaBoostClassifier(DecisionTreeClassifier(), learning_rate=0.1),
    RandomForestClassifier(),
    ExtraTreesClassifier(),
    GradientBoostingClassifier(),
    MLPClassifier(max_iter=1000),
    KNeighborsClassifier(),
    LogisticRegression(max_iter=1000),
    LinearDiscriminantAnalysis(),
    BernoulliNB(),
    GaussianNB()]
```

#### 1.1) Algorithms evaulated on BERT-embeddings

```python
x_train, y_train, x_val, y_val = task1_load_cases(feature="emb", shuffle=True)
evaluate_algorithms(classifiers, x_train, y_train, x_val, y_val, title="Validation scores, task 1 embeddings")
```

#### 1.2) Algorithms evaulated on text features

```python
x_train, y_train, x_val, y_val = task1_load_cases(feature="textf", shuffle=True)
evaluate_algorithms(classifiers, x_train, y_train, x_val, y_val, title="Validation scores, task 1 text features")
```
